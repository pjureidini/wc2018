# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# display preferences
%matplotlib inline
#from IPython.core.display import display, HTML
#display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
def plot_outcomes(df, alpha, emphasis):
    
    fig , ax = plt.subplots(figsize = (10,10))
    
    ax.scatter(df[df['outcome_num'] == 0]['home_rank_pts'],\
               df[df['outcome_num'] == 0]['away_rank_pts'],\
                label = 'Win',\
                alpha = [1 if emphasis == 0 else alpha][0], \
                color = 'blue')

    ax.scatter(df[df['outcome_num'] == 1]['home_rank_pts'],\
               df[df['outcome_num'] == 1]['away_rank_pts'],\
                label = 'Draw',\
                alpha = [1 if emphasis == 1 else alpha][0], \
                color = 'red')
    
    ax.scatter(df[df['outcome_num'] == 2]['home_rank_pts'],\
               df[df['outcome_num'] == 2]['away_rank_pts'],\
                label = 'Loss',\
                alpha = [1 if emphasis == 2 else alpha][0], \
                color = 'green')

    ax.tick_params(axis = 'both', which = 'major', labelsize = 14)
    plt.xlabel('Home Rank (points)', fontsize = 20)
    plt.ylabel('Away Rank (points)', fontsize = 20)

    plt.title('Historical Match Outcomes: 2010 - Present', fontsize = 20)
    #plt.legend(loc = 'lower right')
    plt.legend(fontsize=20)
    #plt.show()
    
    return fig, ax
```

## 1 - General Observations

### 1.1 International  Results

```python
df = pd.read_csv("data/team/new_intl_results.csv")
df['date'] = pd.to_datetime(df['date'])
df['year'] = pd.DatetimeIndex(df['date']).year
```

```python
df.describe()
```

The international results dataset contains almost 40,000 matches, going back to 1872.<br/>
Interestingly, the table above shows that there are extreme scores in the dataset.  The highest home score is 31, and the highest away score is 21!  These numbers are hard to believe, but a [cursory check](https://en.wikipedia.org/wiki/Australia_31â€“0_American_Samoa) shows that they do in fact happen from time to time.

### 1.2 - Friendly Matches

```python
# create yearly bins for plotting
year_bins = np.linspace(min(df.year),max(df.year), max(df.year) - min(df.year) + 1)

fig, ax = plt.subplots(1,1,figsize = (10,5))
ax.hist(df['year'], bins = year_bins, alpha = 0.2, label = 'All macthes')
ax.hist(df[df['tournament'] == 'Friendly']['year'], bins = year_bins, alpha = 0.2, label = 'Friendly')
ax.set_xlabel('Year')
ax.set_ylabel('Frequency')
ax.set_title('Number of Matches per Year')
ax.legend()
plt.show()
```

A high percentage of matches fall in the 'Friendly' category.  We will have to make a decision in terms of how to include these matches in our feature engineering process.  An argument could be made to ignore these matches because the stakes are lower, but discarding them would result in a fairly large loss of data.



### 1.3 - Neutral Site Matches

```python
fig, ax = plt.subplots(1,1,figsize = (10,5))
ax.hist(df['year'], bins = year_bins, alpha = 0.2, label = 'All macthes')
ax.hist(df[df['neutral'] == 1]['year'], bins = year_bins, alpha = 0.2, label = 'Neutral')
ax.set_xlabel('Year')
ax.set_ylabel('Frequency')
ax.set_title('Number of Matches per Year')
ax.legend()
plt.show()
```

Most matches are hosted by one of the teams.  Neutral field matches currently represent about 25% of matches played.  

## 2 - FIFA Rankings vs. Match Outcomes

```python
prj_df = pd.read_csv('data/team/project_data.csv')
```

```python
fig , ax = plot_outcomes(prj_df,0.1,-2) 
plt.show()
```


The plot above provides several insights for our baseline model:<br/>
 - There quite a bit of overlap between the three classes (W,D,L) even though it is apparent that teams with a much higher FIFA rank tend to defeat teams ranked much lower.
 - Imbalanced contests typically have predictable outcomes, but the picture is muddier as the teams get closer from a ranking perspective
 - This plot is a good illustration of the challenge in predicting match results.

## 3 - A Closer Look at our Predictors

### 3.1 - Distributions vs Outcome

```python
final_df = pd.read_csv('data/team/final_dataset.csv', index_col='Unnamed: 0')
```

```python
final_df['date'] = pd.to_datetime(final_df['date'])
```

```python
# exclude world cup results from analysis
pre_wc_df = final_df[final_df['date'] <= '2018-06-13']
```

```python
cols = ['ATT_diff',     'MID_diff',     'DEF_diff',  
        'ATT_age_diff', 'MID_age_diff', 'DEF_age_diff', 
        'ATT_ht_diff',  'MID_ht_diff',  'DEF_ht_diff', 
        'ATT_wt_diff',  'MID_wt_diff',  'DEF_wt_diff',  
        'diff_hrva_rwma', 'diff_hrvap_rwma', 'diff_hrva_rmean' , 'diff_hrvap_rmean', 
        'hrvo_rwma' , 'hrvop_rwma',
        'rank_pts_diff'         
        ]
```

```python
pos = ['W','D','L']


fig, axs = plt.subplots(7,3, figsize = (20,40))#, sharey=True)
#fig.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=.2, hspace=0.5)

for col, ax in zip(cols,axs.ravel()):
    #los = lw.lowess(train_data['outcome_num'], train_data[col], frac = 0.3)
    #print(los)
    # now make the plots
    ax.violinplot(pre_wc_df[col][pre_wc_df['outcome'] == 'W'], [0] , vert = True, showmeans=True) 
    ax.violinplot(pre_wc_df[col][pre_wc_df['outcome'] == 'D'], [1] , vert = True, showmeans=True) 
    ax.violinplot(pre_wc_df[col][pre_wc_df['outcome'] == 'L'], [2] , vert = True, showmeans=True) 
    ax.set_xticks([y for y in range(3)])
    ax.set_xlabel('Match Outcome')
    ax.set_ylabel(col)
    ax.set_title('Match Outcome vs.' + col)
    
    #ax.setp(xticks=['Win','Draw','Loss'])
    
axs[-1, -1].axis('off')
axs[-1, -2].axis('off')
fig.suptitle('Match Outcome vs. Model Variables', fontsize= 20, y = 0.91)

plt.setp(axs,
         xticklabels=['Win','Draw','Loss'])
plt.show()
```

The figure above illustrates how our variables are distributed for wins, draws and losses.  It appears that a good number of variables are not very distinct from each other when broken out into outcomes.  The age related variables are a good example.  They have very similar distributions and means.  This may indicate that they are unlikely to benefit our model.

### 3.2 - Correlations

```python
scatter_matrix = pd.plotting.scatter_matrix(pre_wc_df[cols],
    figsize  = [20, 20],
    marker   = ".",
    s        = 1,
    diagonal = "kde",
    alpha = .5
)

for ax in scatter_matrix.ravel():
    ax.set_xlabel(ax.get_xlabel(), fontsize = 12, rotation = 90)
    ax.set_ylabel(ax.get_ylabel(), fontsize = 12, rotation = 0,ha="right")
    
#plt.setp( ax.xaxis.get_majorticklabels(), rotation=-45, ha="left" )
plt.suptitle('Variable Correclations', fontsize= 20, y = 0.91)
plt.show()
```

Unsurprisingly, several features in the dataset are highly correlated with each other: 
 - diff_hrva_rwma, diff_hrvap_rwma, , diff_hrva_rmean, diff_hrvap_rmean were developed in concert and are variations of the Historical Record Verus All variable
 - hrvo_rwma and hrvop_rwma are also variations of the Historical Record Versus Opponent variable
The model should contain at most one variable from the first set, and one variable from the latter.

Additionaly, we can see that height and weight related variables are somewhat correlated.

### 3.3 - Additional Visualizations

```python
import statsmodels.nonparametric.smoothers_lowess as lw

fig, axs = plt.subplots(7,3, figsize = (20,30), sharey=True)
#fig.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=.2, hspace=0.5)

for col, ax in zip(cols,axs.ravel()):
    los = lw.lowess(pre_wc_df['outcome_num'], pre_wc_df[col], frac = 0.3)
    #print(los)
    # now make the plots
    ax.scatter(x = pre_wc_df[col], y = pre_wc_df['outcome_num'], alpha = 0.05, marker = 's' )
    ax.plot(los[:, 0], los[:, 1], color = 'orange')
    ax.set_xlabel(col)
    ax.set_ylabel('Match Outcome')
    ax.set_title('Match Outcome vs.' + col)
    
axs[-1, -1].axis('off')
axs[-1, -2].axis('off')
fig.suptitle('Match Outcome vs. Variables', fontsize= 20, y = 0.91)
plt.show()
```

Similar to 3.1 above, the figure above provides a hint of the relationship between variables and outcomes.  Age, height and weight-related predictors visually appear weaker than some of the othe variables.
