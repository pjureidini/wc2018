```python
# Get ESPN Scoreboard details
score_board_url = 'http://www.espn.com/soccer/scoreboard/_/league/all/date/20151008'
score_board_soup = BeautifulSoup(requests.get(score_board_url).text, "html.parser")
```

```python
pattern = re.compile("window.espn.scoreboardData 	= {.*")
score_board_html_scr = pattern.findall(score_board_soup.text)
```

```python
score_board_html_scr = str(score_board_html_scr).split("[\'window.espn.scoreboardData \\t= ")[1].split(";window.espn.scoreboardSettings")[0]
```

```python
scores = json.loads(score_board_html_scr)
```

```python
# Loop through matches and store match data in match_date_df
match_date_dict = {}

for score in scores['scores']:
    for event in score['events']:
        match_date_dict[event['name']] = event['uid'].split(':')[3]

match_date_df = pd.DataFrame.from_dict(match_date_dict, orient = 'index', columns = ['uid']).reset_index()

match_date_df['date'] = '20151008'

match_date_df
```
